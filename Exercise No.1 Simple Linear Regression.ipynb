```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('canada_per_capita_income.csv', encoding='latin1')
data
```

```python
data=data.rename({'per capita income (US$)':'income'},axis=1)
data
```

```python
plt.xlabel('year')
plt.ylabel('income')

plt.scatter(data.year,data.income,color='red',marker='*')
```

```python
new_df=data.drop('income',axis=1)
new_df
```

```python
income = data.income
income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
